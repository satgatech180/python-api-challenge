```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy import stats
from scipy.stats import linregress

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# # Save config information.
# url = "http://api.openweathermap.org/data/2.5/weather?"
# units = "imperial"

# # Build partial query URL
# query_url = f"{url}appid={weather_api_key}&units={units}&q="
# query_url

# Starting URL for Weather Map API Call
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key
url
```

```python
#Set up lists to hold weather response info
#Set up lists to hold weather response info
city_name = []
cloudiness = []
country = []
date = []
humidity = []
lat = []
lng = []
clouds = []
max_temp = []
wind_speed = []

#Begin record counter
record = 1

#Begin printing log of each city as it'sbeing processed (with the city number and city name) 
#Not all calls to OpenWeatherMap API will have records in all cities generated by CityPy PIP
print(f'Beginning Data Retrieval')
print(f'----------------------------------') 

i = 0
j = 0

for city in cities:
    
    try:
        response = requests.get(f"{url}&q={city}").json() 
        city_name.append(response['name'])
        cloudiness.append(response['clouds']['all'])
        country.append(response['sys']['country'])
        date.append(response['dt'])
        humidity.append(response['main']['humidity'])
        lat.append(response['coord']['lat'])
        lng.append(response['coord']['lon'])
        max_temp.append(response['main']['temp'])
        wind_speed.append(response['wind']['speed'])
        city_record = response['name']
        print(f"Processing Record {record} of Set 1] | {city_record}")
        i = i +1
        record= record + 1
        
        time.sleep(1.05)
        
    except:
        print("City not found. Skipping..." + city)
        j = j +1
        
    continue

print(" Cities Found =  ",+i)
print(" Cities Skipped =  ",+j)
    
    
    
    
    
    
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
weatherpy_dict = {"City" : city_name,
                   "Cloudiness" : cloudiness,
                   "Country" : country,
                   "Date" : date,
                   "Humidity" : humidity,
                   "Lat" : lat,
                   "Lng" : lng,
                   "Max Temp" : max_temp,
                   "Wind Speed" : wind_speed}

weather_data = pd.DataFrame(weatherpy_dict)

weather_data.count()
```

```python
weather_data.dropna(axis=0, how="any", inplace=True)
weather_data.to_csv("../output_data/cities.csv")
weather_data_csv = "../output_data/cities.csv"
weather_data_df = pd.read_csv(weather_data_csv, encoding ="UTF-8")
```

```python
cityW=weather_data_df.drop(columns="Unnamed: 0")
cityW.to_csv("../output_data/cities.csv")
cityW.head()
```

## Linear Regression

```python
# OPTIONAL: Create a function to create Linear Regression plots
```

```python
# Create Northern and Southern Hemisphere DataFrames
temp_df = cityW[["City", "Lat", "Max Temp"]]

NHlat = temp_df.loc[temp_df["Lat"] >= 0]         

SHlat = temp_df.loc[temp_df["Lat"] < 0]    
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
plt.scatter(NHlat["Lat"], NHlat["Max Temp"])
plt.xlabel("Latitude")
plt.ylabel("Temperature")


(slope, intercept, rvalue, pvalue, stderr) = stats.linregress(NHlat["Lat"], NHlat["Max Temp"])
NHtempregress = NHlat["Lat"] * slope + intercept

line_eq = "y = " + str(round(slope,2)) + "x +" + str(round(intercept,2))
plt.plot(NHlat["Lat"],NHtempregress,"r-")
plt.annotate(line_eq,(20,15),fontsize=15,color="red")

print(f"The r-squared is: {rvalue}")

plt.savefig("../output_data/NHem Temp vs. Lat.png")
plt.show()

```

####  Southern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
plt.scatter(SHlat["Lat"], SHlat["Max Temp"])
plt.xlabel("Latitude")
plt.ylabel("Temperature")


(slope, intercept, rvalue, pvalue, stderr) = stats.linregress(SHlat["Lat"], SHlat["Max Temp"])
SHtempregress = SHlat["Lat"] * slope + intercept

line_eq = "y = " + str(round(slope,2)) + "x +" + str(round(intercept,2))
plt.plot(SHlat["Lat"],SHtempregress,"r-")
plt.annotate(line_eq,(20,15),fontsize=15,color="red")

print(f"The r-squared is: {rvalue}")
plt.savefig("../output_data/SHem Temp vs. Lat.png")
plt.show()
```
